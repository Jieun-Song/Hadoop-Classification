```python
#아마 주소를 제대로 가지고오지 못하는 상태인듯.
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import csv
import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url = "https://xn--939au0g4vj8sq.net/cp/?ca=2005&loca_prt=%EC%84%9C%EC%9A%B8&local_1=%EC%A0%84%EC%B2%B4&local_2=%EC%84%9C%EC%9A%B8"
driver.get(url)
result_list = []

for page in range(0,81):
    if page == 0:
        pass
    else :
        print(page)
        try :
            for i in range((page-1)*20+1,(page-1)*20+21):
                element = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//*[@id="gall_ul"]/li[{}]/div/div[2]/dl/dt/a'.format(i))))
                print(i, element.text, element.get_attribute('href'))
                name = element.text
                href = element.get_attribute('href')
                result_list.append(
                    {
                        "name" : name,
                        "link" : href,
                    }
                )
        except Exception as e:
            print(f"데이터를 가져오는 중 오류 발생: {e}")
            break
            
    SCROLL_PAUSE_TIME = 2

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")
    # 마지막 시점의 창 높이 저장

    while True:
        # 창 높이까지 스크롤
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # 스크롤 후 창이 로딩될때까지 2초를 기다림, 로딩이 다 되면 넘어감
        time.sleep(SCROLL_PAUSE_TIME)
        # 한 번에 맨 마지막까지 스크롤되면 아래 리스트가 뜨지 않아서, 마지막을 찍고 조금 창을 올리는 방법으로 리스트가 로딩될 수 있게 함
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight-50);")
        time.sleep(SCROLL_PAUSE_TIME)

        # 스크롤이 된 후의 창 높이를 새로운 높이로 저장
        new_height = driver.execute_script("return document.body.scrollHeight")

        # 새로운 높이가 이전 높이와 변하지 않았으면 스크롤 종료
        if new_height == last_height:
            break
        
        last_height = new_height

with open ('./gangnamOutput.csv', mode='w') as gangnamOutput:
    brand_writer = csv.writer(gangnamOutput)
    brand_writer.writerow(["name", "link"])
    for a in result_list:
        brand_writer.writerow([a["name"], a["link"]])

        
print("CSV 저장 완료")
# driver.quit()
```

```python
import csv
import requests
from bs4 import BeautifulSoup
import time

f = open('gangnamOutput.csv', "r")
reader = csv.reader(f)
data = list(reader)
result_list = []

success_urls = []
failed_urls = []

for i in data:
    if i[0] == "name":
        continue
    url = i[1]
    
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        failed_urls.append(url)
        continue
    
    if response.status_code == 200:
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')
        
        # 키워드 가져오기
        keyword = soup.select('#key_result > span')
        if not keyword:  # 대체 선택자 시도
            keyword = soup.select('.alternative-keyword-selector')
        
        if len(keyword) > 0:
            success_urls.append(url)
            for a in keyword:
                result_list.append(a.get_text())
        else:
            print(f"No keywords found for {url}")
            failed_urls.append(url)
    else:
        print(f"Failed to fetch {url}, Status Code: {response.status_code}")
        failed_urls.append(url)
    
    time.sleep(1)  # 요청 사이 대기

# CSV로 저장
with open('./gangnamKeyword1.csv', mode='w', newline='', encoding='utf-8') as gangnamKeyword:
    brand_writer = csv.writer(gangnamKeyword)
    brand_writer.writerow(["keyword"])
    for b in result_list:
        brand_writer.writerow([b])

# 성공 및 실패 URL 저장
with open('./successful_urls.txt', 'w') as success_file:
    success_file.writelines("\n".join(success_urls))

with open('./failed_urls.txt', 'w') as failed_file:
    failed_file.writelines("\n".join(failed_urls))

print("끝")
```
