```python
import time
import csv
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

def scroll_down(driver, n):
    for _ in range(n):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)

# WebDriver 설정
options = webdriver.ChromeOptions()
options.add_argument("window-size=1920x1080")
options.add_argument("--disable-blink-features=AutomationControlled")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# URL 접근
url = "https://dinnerqueen.net/taste?ct=%EB%A7%9B%EC%A7%91&area1=%EC%84%9C%EC%9A%B8"
driver.get(url)

# 스크롤 다운 실행
scroll_down(driver, 10)

# 데이터 추출
data = []
parent_selector = "#taste_list > div"  # 부모 요소
parent_elements = driver.find_elements(By.CSS_SELECTOR, parent_selector)

for idx, parent in enumerate(parent_elements, start=1):  # 모든 자식 순회
    child_selector = f"#taste_list > div:nth-child({idx}) > div > a"
    elements = driver.find_elements(By.CSS_SELECTOR, child_selector)
    for element in elements:
        href = element.get_attribute('href')
        title = element.get_attribute('title')
        if href and title:  # href와 title이 모두 있을 경우만 저장
            data.append([title, href])

# CSV 파일로 저장
output_file = "taste_list.csv"
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "URL"])  # 헤더 작성
    writer.writerows(data)

print(f"크롤링 완료. {output_file}에 저장되었습니다.")

# 드라이버 종료
driver.quit()
```

```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

# WebDriver 설정
options = webdriver.ChromeOptions()
options.add_argument("window-size=1920x1080")
options.add_argument("--disable-blink-features=AutomationControlled")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# CSV 파일에서 URL 읽기
input_file = "taste_list.csv"  # 기존 파일 이름
output_file = "filtered_keywords.csv"  # 결과를 저장할 새로운 파일 이름

keywords_only = []

# 입력 파일 읽기 및 데이터 크롤링
with open(input_file, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        url = row["URL"]
        driver.get(url)
        time.sleep(1)  # 페이지 로드 대기

        # MainKeyword에서 텍스트 추출
        try:
            keyword_element = driver.find_element(By.CSS_SELECTOR, "#MainKeyword")
            raw_text = keyword_element.text
            # `,&nbsp;`와 공백 제거
            clean_text = raw_text.replace(",&nbsp;", "").replace("\n", "").strip()
            # '해당 방문 지역 + 맛집 키워드' 필터링
            if "해당 방문 지역 + 맛집 키워드" not in clean_text:
                keywords_only.append([clean_text])
                print(clean_text)
        except Exception as e:
            print(f"키워드 추출 실패: {url}, 오류: {e}")

# 결과를 CSV로 저장
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Keyword"])  # 헤더 작성
    writer.writerows(keywords_only)

print(f"필터링 및 추출 완료. 결과는 {output_file}에 저장되었습니다.")

# 드라이버 종료
driver.quit()
```

```python
import csv

# 입력 및 출력 파일 경로
input_file = "filtered_keywords.csv"
output_file = "processed_keywords.csv"

# 데이터를 처리하고 저장
processed_data = []

with open(input_file, mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # 헤더 건너뛰기
    for row in reader:
        if row:  # 비어 있지 않은 행만 처리
            keywords = row[0].replace('"', '').strip()  # 쌍따옴표 제거 및 양쪽 공백 제거
            split_keywords = [keyword.strip() for keyword in keywords.split(',')]  # ',' 기준 분리 및 공백 제거
            processed_data.extend(split_keywords)  # 리스트에 추가

# 새로운 CSV 파일로 저장
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Keyword"])  # 헤더 작성
    for keyword in processed_data:
        writer.writerow([keyword])  # 한 줄씩 저장

print(f"데이터 처리 완료. 결과는 {output_file}에 저장되었습니다.")
```

```python
import csv
import os

# 입력 파일 경로
input_file1 = "/Users/junho/Hadoop-Classification/data/crawling/gangnamKeyword-modify.csv"
input_file2 = "processed_keywords.csv"

# 출력 파일 경로
output_dir = "data/crawling/ad-keyword"
output_file = os.path.join(output_dir, "ad_keyword_merged.csv")

# 디렉토리 생성
os.makedirs(output_dir, exist_ok=True)

# 데이터를 읽고 결합
merged_data = []

# 첫 번째 파일 읽기
with open(input_file1, mode='r', encoding='utf-8') as file1:
    reader1 = csv.reader(file1)
    header1 = next(reader1)  # 헤더 읽기
    for row in reader1:
        if row:  # 비어 있지 않은 행만 추가
            merged_data.append(row)

# 두 번째 파일 읽기
with open(input_file2, mode='r', encoding='utf-8') as file2:
    reader2 = csv.reader(file2)
    header2 = next(reader2)  # 헤더 읽기
    for row in reader2:
        if row:  # 비어 있지 않은 행만 추가
            merged_data.append(row)

# 결과를 새로운 CSV 파일로 저장
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Keyword"])  # 헤더 작성
    writer.writerows(merged_data)

print(f"파일 결합 완료. 결과는 {output_file}에 저장되었습니다.")
```
