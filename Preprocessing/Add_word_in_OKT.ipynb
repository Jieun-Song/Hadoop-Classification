```python
import os
import re
import json
import time
import requests
import numpy as np 
import pandas as pd
from konlpy.tag import Okt
STOPWORDS_URL = "https://gist.githubusercontent.com/Nine1ll/3ec361da95ccd39101051ab57f86cf46/raw/a1a451421097fa9a93179cb1f1f0dc392f1f9da9/stopwords.txt"

okt = Okt()
drop_pos = ["Josa", "Punctuation", "Foreign", "KoreanParticle"]

def load_stopwords(filepath="Preprocessing/data/stopwords.txt", github_url=STOPWORDS_URL):
        stopwords = []
        if not os.path.exists(filepath):
            if github_url is None:
                raise ValueError("GitHub url이 제공되지 않았습니다.")
            
            # 다운로드하기
            response = requests.get(github_url)
            if response.status_code == 200:
                # 멀쩡하게 반응하면
                os.makedirs(os.path.dirname(filepath), exist_ok=True)
                with open(filepath, "w", encoding="utf-8") as f:
                    f.write(response.text)
                print(f"stopwords.txt를 {github_url}에서 {filepath}에 다운로드 했습니다. ")
            else:
                raise Exception(f"우와 에러다. HTTP Status: {response.status_code} 실패했습니다.")
        
        # 파일 읽어오기
        with open(filepath, "r", encoding="utf-8") as f:
            stopwords = [line.strip() for line in f.readlines()]
        
        return stopwords

def preprocessing(text, norm=False, stem=False):
    stopwords = load_stopwords()
    text = str(text)
    text = regex(text)
    words = okt.pos(text,norm=norm, stem=stem)
    words = [word for word, pos in words if pos not in drop_pos]
    words = [regex(word) for word in words if word not in stopwords and regex(word).strip()]
    
    return " ".join(words)


def regex(text):
    """
    거슬리는 데이터들 삭제
    정규 표현식 패턴 (인터넷 주소와 "© NAVER Corp" 삭제)
    한 단어 이하 삭제
    공백 제거
    시간 형식 삭제
    """
    print(text+"\n")
    pattern = r"(© NAVER Corp|https?://\S+|www\.\S+|m\.\S+|\S+\.\S+\.\S+)"
    cleaned_text = re.sub(pattern, "", text)
    cleaned_text = re.sub(r"\b\d{3,4}-\d{3,4}-\d{3,4}\b", "", cleaned_text)# 전화번호 형식 삭제 (예: 0507-1311, 010-1234-5678)
    cleaned_text = re.sub(r"\b\d+\b", "", cleaned_text)  # 숫자만 있는 부분을 제거
    cleaned_text = re.sub(r"\b\d{1,2}:\d{2}\b", "", cleaned_text) # 시간 형식 제외
    cleaned_text = re.sub(r"\b\w\b", "", cleaned_text) # 공백 제거
    cleaned_text = re.sub(r"\s+", " ", cleaned_text).strip() # 한 단어 이하 삭제


df = pd.read_csv("/Users/nine1ll/BDP_termProject/Hadoop-Classification/data/crawling/blog-review/양천_blog_review.csv")
print(df.shape)
df.dropna(inplace=True)
print(df.shape)
```
